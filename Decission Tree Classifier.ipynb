```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
dataset.target
```

```python
X=df.iloc[:,:-1]
y=dataset.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(criterion='entropy')
```

```python
clf
```

```python
clf.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,12))
tree.plot_tree(clf,filled=True)
```

## post purning technique

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=3)
clf.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,12))
tree.plot_tree(clf,filled=True)
```

```python
## prediction
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

## Pre purning and hyper parameter turning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=DecisionTreeClassifier()
clff=GridSearchCV(clf,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clff.fit(X_train,y_train)
```

```python
clff.best_params_
```

```python
y_predd=clff.predict(X_test)
```

```python
y_predd
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_predd,y_test)
print(score)
print(classification_report(y_predd,y_test))
```
