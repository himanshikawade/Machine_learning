K means Clustering  algorithmn 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
#3 use elbow method to select the k value
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
    
```

```python
wcss
```

```python
# plot the elbow curve

plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel('WCSS')
plt.show()
```

```python
## Silhoutte Score(performance matrix)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficient=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    score=silhouette_score(X_train,kmeans.labels_)
    silhouette_coefficient.append(score)
```

```python
silhouette_coefficient
```

```python
# plot the siloutte score

plt.plot(range(2,11),silhouette_coefficient)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel('Silhoutee Coefficient')
plt.show()
```
